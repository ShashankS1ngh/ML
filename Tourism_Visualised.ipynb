```python
# 1. Heatmap of quarterly visitors by continent
quarter_data = pd.read_excel('VAA Final Dataset.xlsx', sheet_name='Country Quarter Wise Visitors')

# Group countries by continent (example grouping)
continent_mapping = {
    'U.S.A': 'North America', 'Canada': 'North America',
    'U.K.': 'Europe', 'France': 'Europe', 'Germany': 'Europe',
    'China': 'Asia', 'Japan': 'Asia', 'Malaysia': 'Asia',
    'Australia': 'Oceania',
    'Brazil': 'South America', 'Argentina': 'South America'
}

quarter_data['Continent'] = quarter_data['Country of Nationality'].map(continent_mapping)
quarter_cols = [col for col in quarter_data.columns if 'quarter' in col]
continent_quarters = quarter_data.groupby('Continent')[quarter_cols].mean()

plt.figure(figsize=(15, 8))
sns.heatmap(continent_quarters, annot=True, fmt='.0f', cmap='YlOrRd')
plt.title('Average Quarterly Visitors by Continent')
plt.xlabel('Quarter')
plt.ylabel('Continent')
plt.xticks(rotation=45)
plt.show()

# 2. Year-over-year growth rate for monthly visitors
monthly_visitors_pivot = monthly_visitors.set_index('Year')
yoy_growth = monthly_visitors_pivot.pct_change() * 100

plt.figure(figsize=(15, 8))
sns.boxplot(data=yoy_growth)
plt.title('Year-over-Year Growth Rate Distribution in Monthly Visitors')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.xticks(rotation=45)
plt.grid(True, axis='y')
plt.show()

# 3. Seasonal decomposition of monthly visitors
from statsmodels.tsa.seasonal import seasonal_decompose
import numpy as np

# Calculate total monthly visitors
monthly_visitors_total = monthly_visitors.set_index('Year').mean(axis=0)
result = seasonal_decompose(monthly_visitors_total, period=12)

fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(15, 12))
result.observed.plot(ax=ax1)
ax1.set_title('Observed')
ax1.grid(True)
result.trend.plot(ax=ax2)
ax2.set_title('Trend')
ax2.grid(True)
result.seasonal.plot(ax=ax3)
ax3.set_title('Seasonal')
ax3.grid(True)
result.resid.plot(ax=ax4)
ax4.set_title('Residual')
ax4.grid(True)
plt.tight_layout()
plt.show()

# 4. Correlation between FEE and Visitors by month
monthly_fee_pivot = monthly_fee.set_index('Year')
correlation_by_month = pd.DataFrame()

for year in monthly_visitors_pivot.columns:
    if year in monthly_fee_pivot.columns:
        correlation_by_month[year] = [monthly_visitors_pivot[year].corr(monthly_fee_pivot[year])]

plt.figure(figsize=(12, 6))
sns.barplot(data=correlation_by_month.T)
plt.title('Correlation between Monthly Visitors and FEE by Year')
plt.xlabel('Year')
plt.ylabel('Correlation Coefficient')
plt.grid(True, axis='y')
plt.show()

# 5. Monthly distribution of visitors using violin plots
monthly_visitors_melt = pd.melt(monthly_visitors, id_vars=['Year'], var_name='Year_num', value_name='Visitors')
plt.figure(figsize=(15, 8))
sns.violinplot(data=monthly_visitors_melt, x='Year_num', y='Visitors')
plt.title('Distribution of Monthly Visitors by Year')
plt.xlabel('Year')
plt.ylabel('Number of Visitors')
plt.xticks(rotation=45)
plt.grid(True, axis='y')
plt.show()
```

```python


# Load the data
monthly_visitors = pd.read_excel('VAA Final Dataset.xlsx', sheet_name='Monthly Visitors Data')
monthly_fee = pd.read_excel('VAA Final Dataset.xlsx', sheet_name='Monthly FEE in Crore')

# 1. Monthly FEE per Visitor Analysis
monthly_visitors_pivot = monthly_visitors.set_index('Year')
monthly_fee_pivot = monthly_fee.set_index('Year')

# Calculate average FEE per month
avg_monthly_fee = monthly_fee_pivot.mean()
avg_monthly_visitors = monthly_visitors_pivot.mean()
fee_per_visitor = avg_monthly_fee / avg_monthly_visitors

plt.figure(figsize=(15, 8))
fee_per_visitor.plot(kind='bar')
plt.title('Average FEE per Visitor by Month')
plt.xlabel('Month')
plt.ylabel('FEE per Visitor (Crore/Visitor)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# 2. Monthly Revenue Patterns
plt.figure(figsize=(15, 8))
monthly_fee_pivot.mean().plot(marker='o', linewidth=2, markersize=10)
plt.fill_between(range(len(monthly_fee_pivot.mean())),
                monthly_fee_pivot.mean() - monthly_fee_pivot.std(),
                monthly_fee_pivot.mean() + monthly_fee_pivot.std(),
                alpha=0.3)
plt.title('Average Monthly Foreign Exchange Earnings with Variation')
plt.xlabel('Month')
plt.ylabel('FEE (in Crore)')
plt.grid(True)
plt.show()

# 3. Seasonal Revenue vs Visitor Volume
fig, ax1 = plt.subplots(figsize=(15, 8))

# Plot visitors
color = 'tab:blue'
ax1.set_xlabel('Month')
ax1.set_ylabel('Average Visitors', color=color)
ax1.plot(monthly_visitors_pivot.mean().values, color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Plot FEE on secondary axis
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Average FEE (Crore)', color=color)
ax2.plot(monthly_fee_pivot.mean().values, color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Monthly Comparison: Visitors vs Foreign Exchange Earnings')
plt.grid(True)
plt.show()

# 4. Year-over-Year Growth Analysis
yearly_total_fee = monthly_fee_pivot.sum()
yearly_total_visitors = monthly_visitors_pivot.sum()

plt.figure(figsize=(15, 8))
width = 0.35
x = np.arange(len(yearly_total_fee))
plt.bar(x - width/2, yearly_total_fee/100000, width, label='FEE (Lakh Crore)')
plt.bar(x + width/2, yearly_total_visitors/1000000, width, label='Visitors (Millions)')
plt.title('Yearly Comparison: Total FEE vs Total Visitors')
plt.xlabel('Year')
plt.legend()
plt.grid(True)
plt.show()

# Print key insights
print("\
Key Insights:")
print("1. Months with highest FEE per visitor:",
      fee_per_visitor.nlargest(3).index.tolist())
print("2. Peak revenue months:",
      monthly_fee_pivot.mean().nlargest(3).index.tolist())
print("3. Average FEE per visitor (in Crore):",
      round(fee_per_visitor.mean(), 4))
print("4. Months with highest visitor variability:",
      monthly_visitors_pivot.std().nlargest(3).index.tolist())
```
